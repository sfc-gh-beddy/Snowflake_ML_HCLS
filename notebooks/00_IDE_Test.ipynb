```python
# Test 1: Basic IDE functionality
print("Testing IDE Integration...")
print("SUCCESS: Notebook cell executed successfully!")
print("Python version check...")

import sys
print(f"   Python: {sys.version}")
print(f"   Environment: {sys.prefix}")

# Check if we're in the right environment
if 'snowflake-ml-platform' in sys.prefix:
    print("SUCCESS: Correct conda environment detected")
else:
    print("WARNING: May not be in the snowflake-ml-platform environment")
    print("   Make sure you selected the right kernel!")
```

```python
# Test 2: Package imports
print("Testing package imports...")

try:
    import snowflake.snowpark
    print("SUCCESS: snowflake-snowpark-python")
except ImportError as e:
    print(f"FAILED: snowflake-snowpark-python: {e}")

try:
    import snowflake.ml
    print("SUCCESS: snowflake-ml-python")
except ImportError as e:
    print(f"FAILED: snowflake-ml-python: {e}")

try:
    import pandas as pd
    import numpy as np
    print("SUCCESS: pandas and numpy")
except ImportError as e:
    print(f"FAILED: pandas/numpy: {e}")

try:
    import sklearn
    import xgboost
    print("SUCCESS: scikit-learn and xgboost")
except ImportError as e:
    print(f"FAILED: ML libraries: {e}")

print("\nPackage imports test complete!")
```

```python
# Test 3: Environment variables and connection helper
print("Testing environment configuration...")

try:
    import os
    from dotenv import load_dotenv
    
    # Load environment variables
    load_dotenv()
    
    # Check if credentials are configured
    account = os.getenv('SNOWFLAKE_ACCOUNT')
    user = os.getenv('SNOWFLAKE_USER')
    
    if account and user:
        print("SUCCESS: Environment variables loaded")
        print(f"   Account: {account}")
        print(f"   User: {user}")
    else:
        print("WARNING: Environment variables not found")
        print("   Please create .env file with your Snowflake credentials")
        
except Exception as e:
    print(f"FAILED: Environment error: {e}")

# Add path for connection module
import sys
current_dir = os.getcwd()
if "notebooks" in current_dir:
    src_path = os.path.join(current_dir, "..", "src")
else:
    src_path = os.path.join(current_dir, "src")
sys.path.append(src_path)

try:
    from snowflake_connection import get_session
    print("SUCCESS: Connection helper imported successfully")
except ImportError as e:
    print(f"FAILED: Connection helper import failed: {e}")
```

```python
# Test 4: Optional Snowflake connection test
print("Testing Snowflake connection (optional)...")
print("This will only work if you've configured your .env file")

try:
    # Only test connection if credentials are available
    if os.getenv('SNOWFLAKE_ACCOUNT') and os.getenv('SNOWFLAKE_USER'):
        session = get_session()
        
        if session:
            result = session.sql("SELECT CURRENT_VERSION()").collect()
            version = result[0][0]
            print("SUCCESS: Snowflake connection successful!")
            print(f"   Version: {version}")
        else:
            print("FAILED: Connection failed - check your credentials")
    else:
        print("SKIPPED: No credentials configured")
        print("   This is normal if you haven't set up .env yet")
        
except Exception as e:
    print(f"WARNING: Connection test failed: {e}")
    print("   This is normal if you haven't configured Snowflake credentials yet")
```

- Make sure you selected the `snowflake-ml-platform` kernel

- Check that you're in the right environment

**Connection Issues:**
- Create `.env` file: `cp env_template.txt .env`
- Edit `.env` with your Snowflake credentials
- Run full connection test: `python connection_test.py`
